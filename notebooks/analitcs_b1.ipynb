```python
# Função para carregar e filtrar uma aba da planilha
def load_and_filter_sheet(file_path, sheet_name):
    # Carregar a aba específica
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    
    # Filtrar classes com Support > 0
    df_filtered = df[df['Support'] > 0]
    
    # Converter colunas numéricas para float (se necessário)
    numeric_columns = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'Support']
    df_filtered[numeric_columns] = df_filtered[numeric_columns].astype(float)
    
    return df_filtered

```

```python
# Função para comparar duas tabelas
def compare_tables(df_orig, df_comp):
    # Garantir que ambas as tabelas tenham as mesmas classes
    common_classes = df_orig["Classes"].isin(df_comp["Classes"])
    df_orig_filtered = df_orig[common_classes]
    df_comp_filtered = df_comp[df_comp["Classes"].isin(df_orig["Classes"])]

    if df_orig_filtered.empty or df_comp_filtered.empty:
        print("Nenhuma classe em comum encontrada entre as tabelas.")
        return pd.DataFrame()

    # Criar um DataFrame vazio para armazenar os resultados da comparação
    comparison_df = pd.DataFrame()
    comparison_df["Classes"] = df_orig_filtered["Classes"]
    
    # Definir as métricas a serem comparadas
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    
    for metric in metrics:
        orig_metric = f"{metric}_Orig"
        comp_metric = f"{metric}_Comp"
        
        # Adicionar colunas ao DataFrame de comparação
        comparison_df[orig_metric] = df_orig_filtered[metric].values
        comparison_df[comp_metric] = df_comp_filtered[metric].values
        
        # Identificar casos iguais, superiores ou inferiores
        comparison_df[f"{metric}_Comparison"] = "Equal"
        comparison_df.loc[
            comparison_df[orig_metric] < comparison_df[comp_metric], f"{metric}_Comparison"
        ] = "Better"
        comparison_df.loc[
            comparison_df[orig_metric] > comparison_df[comp_metric], f"{metric}_Comparison"
        ] = "Worse"
    
    return comparison_df
```
